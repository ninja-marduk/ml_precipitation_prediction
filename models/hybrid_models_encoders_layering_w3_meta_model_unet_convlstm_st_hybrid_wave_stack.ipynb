```python
# %% [markdown]
# # Prediccin Espaciotemporal de Precipitacin Mensual - Notebook Completo

# %% [markdown]
# ## 0) Configuracin entorno, rutas y dependencias

# %%
import sys
import os
import logging
from pathlib import Path
import joblib # Importamos joblib para persistir los scalers
import datetime # Para manejar fechas para rutas de archivos

logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)
logger = logging.getLogger(__name__)

IN_COLAB = "google.colab" in sys.modules
logger.info(f"Ejecutando en Colab: {IN_COLAB}")

if IN_COLAB:
    from google.colab import drive
    drive.mount('/content/drive', force_remount=True)
    BASE_PATH = Path('/content/drive/MyDrive/ml_precipitation_prediction')
    if not Path('ml_precipitation_prediction').exists():
        logger.info("Clonando repositorio ml_precipitation_prediction")
        !git clone https://github.com/ninja-marduk/ml_precipitation_prediction.git
    %cd ml_precipitation_prediction
    # Actualizado pip install para incluir joblib
    !pip install -q xarray netCDF4 optuna matplotlib seaborn lightgbm xgboost scikit-learn ace_tools_open cartopy geopandas joblib
else:
    # Determine project root for local execution
    current_path = Path.cwd()
    project_root_found = None
    # Traverse up from current_path to find a marker like '.git' or 'requirements.txt'
    # This helps ensure BASE_PATH is the actual project root,
    # even if the notebook is run from a subdirectory.
    for p_dir in [current_path] + list(current_path.parents):
        # Check for common project root markers
        if (p_dir / '.git').is_dir() or \
           (p_dir / 'requirements.txt').is_file() or \
           (p_dir / 'README.md').is_file(): # Added README.md as another common marker
            project_root_found = p_dir
            break
    
    if project_root_found:
        BASE_PATH = project_root_found
    else:
        # Fallback to current working directory if no marker is found
        BASE_PATH = current_path
    logger.info(f"Ejecutando en local. Base path: {BASE_PATH}")

DATA_OUTPUT = BASE_PATH / 'data' / 'output'
MODELS_OUTPUT = BASE_PATH / 'models' / 'output'
PREDS_DIR = MODELS_OUTPUT / 'base_model_predictions' # Directorio para guardar predicciones de modelos base

FULL_NC = DATA_OUTPUT / 'complete_dataset_with_features_with_clusters_elevation_with_windows.nc'
FEATURES_NC = MODELS_OUTPUT / 'features_fusion_branches.nc'
SHP_PATH = BASE_PATH / 'data' / 'input' / 'shapes' / 'MGN_Departamento.shp'

for fpath in [FULL_NC, FEATURES_NC, SHP_PATH]:
    if not fpath.exists():
        logger.warning(f"Archivo no encontrado: {fpath}")
    else:
        logger.info(f"Archivo encontrado: {fpath}")

MODELS_OUTPUT.mkdir(parents=True, exist_ok=True)
PREDS_DIR.mkdir(parents=True, exist_ok=True) # Crear el directorio para guardar predicciones

# Parmetros generales - AUMENTADOS PARA UN MEJOR ENTRENAMIENTO
INPUT_WINDOW   = 60
OUTPUT_HORIZON = 3
BATCH_SIZE     = 16
MAX_EPOCHS     = 300 # Aumentado significativamente
PATIENCE       = 50  # Aumentado para early stopping menos agresivo
LR             = 1e-3 # Tpicamente un buen punto de partida

import torch
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
logger.info(f"Usando dispositivo: {DEVICE}")

# %% [markdown]
# ## 1) Imports adicionales y funciones auxiliares

# %%
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import xarray as xr

def evaluate_metrics_np(y_true, y_pred):
    if y_true.shape != y_pred.shape:
        logger.warning(f"Formas de entrada a evaluate_metrics_np difieren: y_true {y_true.shape}, y_pred {y_pred.shape}. Truncando.")
        min_len = min(y_true.size, y_pred.size)
        y_true = y_true.flatten()[:min_len]
        y_pred = y_pred.flatten()[:min_len]
    
    # Asegurarse de que no haya NaNs en las entradas
    valid_indices = ~np.isnan(y_true) & ~np.isnan(y_pred)
    y_true_clean = y_true[valid_indices]
    y_pred_clean = y_pred[valid_indices]

    if y_true_clean.size == 0:
        return np.nan, np.nan, np.nan, np.nan # No hay datos vlidos para calcular mtricas

    rmse = np.sqrt(np.mean((y_true_clean - y_pred_clean) ** 2))
    mae = np.mean(np.abs(y_true_clean - y_pred_clean))
    
    # Manejo de MAPE: Evitar divisiones por cero o valores indeterminados
    # y_true_clean + 1e-5 para evitar /0, y clip a valores razonables para evitar infinitos.
    mape = np.mean(np.abs((y_true_clean - y_pred_clean) / (y_true_clean + 1e-5))) * 100
    
    # Manejo de R2 y NSE: Si el valor real es constante, el denominador es 0.
    # Adems, si y_true_clean tiene muy poca variabilidad, R2 puede ser inestable.
    y_true_variance = np.sum((y_true_clean - np.mean(y_true_clean)) ** 2)
    if y_true_variance < 1e-6: # Usar un umbral pequeo para considerar "constante"
        r2 = np.nan # R2 es indefinido si y_true es constante
        nse = np.nan # NSE es indefinido si y_true es constante
    else:
        r2 = 1 - np.sum((y_true_clean - y_pred_clean) ** 2) / y_true_variance
        nse = r2 # NSE es igual a R2 para el modelo de regresin simple

    return rmse, mae, mape, r2

def plot_training_history(history, title="Training and Validation Loss"):
    plt.figure(figsize=(8,5))
    plt.plot(history["train_loss"], label="Train Loss")
    plt.plot(history["val_loss"], label="Validation Loss")
    plt.title(title)
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    plt.grid(True)
    plt.show()

# %% [markdown]
# ## 2) Datasets PyTorch

# %%
class BaseDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.from_numpy(X).float()
        self.y = torch.from_numpy(y).float()
    def __len__(self): return len(self.X)
    def __getitem__(self, idx): return self.X[idx], self.y[idx]

class MetaDataset(Dataset):
    def __init__(self, X, y):
        # X and y here are NumPy arrays of paths, so we need to load them lazily
        self.X_paths = X # X is now an array of lists of paths (sample, horizon, model, path)
        self.y_paths = y # y is now an array of lists of paths (sample, horizon, path)
    def __len__(self): return len(self.X_paths)
    def __getitem__(self, idx): 
        # Load X data for this sample
        x_sample_paths = self.X_paths[idx] # list of lists of paths for this sample
        x_data_loaded = []
        for horizon_paths in x_sample_paths: # For each horizon
            horizon_models_data = []
            for model_path in horizon_paths: # For each model in this horizon
                horizon_models_data.append(np.load(model_path)) # Load (H, W) array
            x_data_loaded.append(np.stack(horizon_models_data, axis=0)) # Stack models: (M, H, W)
        x_data_loaded = np.stack(x_data_loaded, axis=0) # Stack horizons: (T_meta, M, H, W)
        
        # Load y data for this sample
        y_sample_paths = self.y_paths[idx] # list of paths for this sample
        y_data_loaded = []
        for y_path in y_sample_paths: # For each horizon
            y_data_loaded.append(np.load(y_path)) # Load (H, W) array
        y_data_loaded = np.stack(y_data_loaded, axis=0) # Stack horizons: (T_meta, H, W)

        # Add channel dimension if needed (meta-model expects (B, T_meta, M, H, W, C) for X and (B, T_meta, H, W, C) for Y)
        # Assuming single channel output/input from base models
        if x_data_loaded.ndim == 4: # (T_meta, M, H, W)
            x_data_loaded = x_data_loaded[:, :, :, :, np.newaxis] # (T_meta, M, H, W, 1)
        if y_data_loaded.ndim == 3: # (T_meta, H, W)
            y_data_loaded = y_data_loaded[:, :, :, np.newaxis] # (T_meta, H, W, 1)

        return torch.from_numpy(x_data_loaded).float(), torch.from_numpy(y_data_loaded).float()


# %% [markdown]
# ## 3) Modelos base: GRUEncoderDecoder y TransformerModel

# %%
# Modelo simplificado GRUEncoderDecoder
class GRUEncoderDecoder(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_horizon, output_dim, dropout=0.0):
        super().__init__()
        self.hidden_dim = hidden_dim
        self.output_horizon = output_horizon
        self.output_dim = output_dim

        self.encoder = nn.GRU(input_dim, hidden_dim, batch_first=True)
        self.decoder = nn.GRU(hidden_dim, hidden_dim, batch_first=True)
        self.fc_out = nn.Linear(hidden_dim, output_dim)
        self.dropout = nn.Dropout(dropout) # Add dropout layer

    def forward(self, x):
        # x: (B, T, F)
        _, h = self.encoder(x)
        dec_input = h.repeat(self.output_horizon, 1, 1).permute(1,0,2)  # (B, output_horizon, hidden_dim)
        dec_out, _ = self.decoder(dec_input)
        out = self.fc_out(self.dropout(dec_out))  # Apply dropout before final FC
        return out

# Modelo simplificado TransformerModel
class TransformerModel(nn.Module):
    def __init__(self, input_dim, output_horizon, output_dim, nhead=4, num_layers=2, dim_feedforward=128, dropout=0.1): # Add dropout
        super().__init__()
        self.output_horizon = output_horizon
        self.output_dim = output_dim
        self.input_proj = nn.Linear(input_dim, dim_feedforward)
        encoder_layer = nn.TransformerEncoderLayer(d_model=dim_feedforward, nhead=nhead, dropout=dropout) # Apply dropout here
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)
        self.fc_out = nn.Linear(dim_feedforward, output_dim * output_horizon)

    def forward(self, x):
        # x: (B, T, F)
        x = self.input_proj(x)
        x = x.permute(1,0,2)  # Transformer espera (S, B, F)
        x = self.transformer(x)
        x = x.mean(dim=0)  # (B, F)
        out = self.fc_out(x)  # (B, output_dim * output_horizon)
        out = out.view(-1, self.output_horizon, self.output_dim)  # (B, output_horizon, output_dim)
        return out

# %% [markdown]
# ## 4) ConvLSTMCell, ConvLSTM y Meta-modelo UNetConvLSTMMeta

# %%
import torch.nn.functional as F

class ConvLSTMCell(nn.Module):
    def __init__(self, in_chan, hid_chan, kernel_size=3):
        super().__init__()
        padding = kernel_size // 2
        self.conv = nn.Conv2d(in_chan + hid_chan, 4 * hid_chan, kernel_size, padding=padding)
        self.hid_chan = hid_chan

    def forward(self, x, h, c):
        combined = torch.cat([x, h], dim=1)
        conv_out = self.conv(combined)
        i, f, o, g = torch.split(conv_out, self.hid_chan, dim=1)
        i = torch.sigmoid(i)
        f = torch.sigmoid(f)
        o = torch.sigmoid(o)
        g = torch.tanh(g)
        c_next = f * c + i * g
        h_next = o * torch.tanh(c_next)
        return h_next, c_next

class ConvLSTM(nn.Module):
    def __init__(self, in_chan, hid_chan, kernel_size=3, num_layers=1):
        super().__init__()
        self.cells = nn.ModuleList([
            ConvLSTMCell(in_chan if i == 0 else hid_chan, hid_chan, kernel_size)
            for i in range(num_layers)
        ])

    def forward(self, x):
        B, T, C, H, W = x.shape
        h = [torch.zeros(B, cell.hid_chan, H, W, device=x.device) for cell in self.cells]
        c = [torch.zeros_like(h_i) for h_i in h]
        outs = []
        for t in range(T):
            u = x[:, t]
            for i, cell in enumerate(self.cells):
                h[i], c[i] = cell(u, h[i], c[i])
                u = h[i]
            outs.append(h[-1])
        return torch.stack(outs, dim=1)

class UNetConvLSTMMeta(nn.Module):
    def __init__(self, num_models, ny, nx, output_horizon, channels=1):
        super().__init__()
        self.num_models = num_models
        self.channels = channels
        self.ny = ny 
        self.nx = nx
        self.out_horizon = output_horizon # This will be 3 for the meta-model now

        in_ch = num_models * channels # Num_models (9) * channels (1)
        self.enc1 = nn.Sequential(
            nn.Conv2d(in_ch, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.Conv2d(64, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
        )
        self.pool1 = nn.MaxPool2d(2)
        self.enc2 = nn.Sequential(
            nn.Conv2d(64, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.Conv2d(128, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
        )
        self.pool2 = nn.MaxPool2d(2)
        
        self.bottle_conv = nn.Conv2d(128, 128, 3, padding=1)
        self.convlstm = ConvLSTM(128, 64, 3, 1) 

        self.up1_conv = nn.ConvTranspose2d(64, 64, kernel_size=2, stride=2)
        
        self.dec1 = nn.Sequential(
            nn.Conv2d(64 + 128, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.Conv2d(128, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
        )
        
        self.up2_conv = nn.ConvTranspose2d(128, 64, kernel_size=2, stride=2, 
                                           output_padding=(self.ny % 2, self.nx % 2))
        
        self.dec2 = nn.Sequential(
            nn.Conv2d(64 + 64, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.Conv2d(64, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
        )
        self.out_conv = nn.Conv2d(64, channels, 1) 

    def forward(self, x):
        # x expected shape: (B, T_meta=OUTPUT_HORIZON, M_models=9, H, W, C=1)
        if x.dim() != 6: 
            raise RuntimeError(f"Input tensor expected 6D (B, T_meta, M, H, W, C), got {x.dim()}D. Shape: {x.shape}")

        B, T_meta, M_models, ny_in, nx_in, ch = x.shape
        
        if T_meta != self.out_horizon:
             raise RuntimeError(f"Input temporal dimension mismatch: expected {self.out_horizon}, got {T_meta}.")

        # Reshape input to (B*T_meta, M*C, H, W) for encoder
        x_reshaped_for_unet = x.permute(0, 1, 5, 2, 3, 4).reshape(B * T_meta, M_models * ch, ny_in, nx_in)

        # Encoder (spatial processing for each B*T_meta slice)
        e1 = self.enc1(x_reshaped_for_unet)    
        p1 = self.pool1(e1)
        e2 = self.enc2(p1)   
        p2 = self.pool2(e2) 
        
        # Bottleneck convolution
        b_conv = self.bottle_conv(p2)
        
        # Reshape for ConvLSTM: (B, T_meta, C_bottleneck, H_spatial_reduced, W_spatial_reduced)
        b_conv_for_lstm = b_conv.reshape(B, T_meta, b_conv.shape[1], b_conv.shape[2], b_conv.shape[3])
        
        # ConvLSTM processes the sequence across the temporal dimension T_meta
        cl_output = self.convlstm(b_conv_for_lstm) 

        # Reshape ConvLSTM output back to (B*T_meta, C_ConvLSTM, H_spatial_reduced, W_spatial_reduced) for decoder
        lf = cl_output.reshape(B * T_meta, cl_output.shape[2], cl_output.shape[3], cl_output.shape[4])
        
        # Decoder
        u1 = self.up1_conv(lf) 
        e2_reshaped = e2 # e2 is already (B*T_meta, 128, H_mid, W_mid)
        if u1.size(2) != e2_reshaped.size(2) or u1.size(3) != e2_reshaped.size(3):
            raise RuntimeError(f"Spatial dimension mismatch after upsample 1: u1 {u1.shape}, e2 {e2_reshaped.shape}. Review up1_conv parameters.")
        
        cat1 = torch.cat([u1, e2_reshaped], dim=1)
        d1 = self.dec1(cat1)
        
        u2 = self.up2_conv(d1) 
        e1_reshaped = e1 # e1 is already (B*T_meta, 64, H_orig, W_orig)
        if u2.size(2) != e1_reshaped.size(2) or u2.size(3) != e1_reshaped.size(3):
            raise RuntimeError(f"Spatial dimension mismatch after upsample 2: u2 {u2.shape}, e1 {e1_reshaped.shape}. Review up2_conv parameters.")
        
        cat2 = torch.cat([u2, e1_reshaped], dim=1)
        d2 = self.dec2(cat2)
        
        # Final output convolution
        out = self.out_conv(d2)  # (B*T_meta, channels, H, W) -> channels=1
        
        # Reshape to final desired output format: (B, T_meta, H, W, C)
        B_final = B
        T_meta_final = T_meta
        H_out = out.shape[2]
        W_out = out.shape[3]
        output = out.reshape(B_final, T_meta_final, self.channels, H_out, W_out).permute(0, 1, 3, 4, 2)
        
        if not self.training:
            logger.info(f"Final output shape of the meta-model: {output.shape}")
        
        return output

# %% [markdown]
# ## 5) Funciones de entrenamiento con Early Stopping

# %%
def train_base_model(model, optimizer, criterion, train_loader, val_loader,
                     max_epochs=MAX_EPOCHS, patience=PATIENCE, save_path=None):
    best_val = float('inf')
    counter = 0
    history = {"train_loss": [], "val_loss": []}
    model.to(DEVICE)
    if save_path is None:
        save_path = MODELS_OUTPUT / "best_base.pt"

    for ep in range(max_epochs):
        model.train()
        train_losses = []
        for Xb, yb in train_loader:
            Xb, yb = Xb.to(DEVICE), yb.to(DEVICE)
            optimizer.zero_grad()
            pred = model(Xb)
            loss = criterion(pred, yb)
            loss.backward()
            optimizer.step()
            train_losses.append(loss.item())

        model.eval()
        val_losses = []
        with torch.no_grad():
            for Xv, yv in val_loader:
                Xv, yv = Xv.to(DEVICE), yv.to(DEVICE)
                loss = criterion(model(Xv), yv)
                val_losses.append(loss.item())

        train_loss_avg = np.mean(train_losses)
        val_loss_avg = np.mean(val_losses)
        history["train_loss"].append(train_loss_avg)
        history["val_loss"].append(val_loss_avg)
        logger.info(f"[Base] Epoch {ep+1}/{max_epochs} - Train Loss: {train_loss_avg:.4f} - Val Loss: {val_loss_avg:.4f}")

        if val_loss_avg < best_val:
            best_val = val_loss_avg
            counter = 0
            torch.save(model.state_dict(), save_path)
        else:
            counter += 1
            if counter > patience:
                logger.info(f"Early stopping activado en modelo base en epoch {ep+1}")
                break

    model.load_state_dict(torch.load(save_path, weights_only=False)) 
    return model, history

def train_meta_model(model, optimizer, criterion, train_loader, val_loader,
                     max_epochs=MAX_EPOCHS, patience=PATIENCE, save_path=None):
    best_val = float('inf')
    counter = 0
    history = {"train_loss": [], "val_loss": []}
    model.to(DEVICE)
    if save_path is None:
        save_path = MODELS_OUTPUT / "best_meta.pt"

    for ep in range(max_epochs):
        model.train()
        train_losses = []
        for Xb, yb in train_loader: # Xb is (B, T_meta, M, H, W, C), yb is (B, T_meta, H, W, C)
            Xb, yb = Xb.to(DEVICE), yb.to(DEVICE)
            optimizer.zero_grad()
            
            if ep == 0 and len(train_losses) == 0:
                logger.info(f"Forma de entrada Xb (meta-modelo): {Xb.shape}")
                logger.info(f"Forma de objetivo yb (meta-modelo, antes de squeeze): {yb.shape}")
            
            try:
                output = model(Xb) # output is (B, T_meta, H, W, C)
                
                if ep == 0 and len(train_losses) == 0:
                    logger.info(f"Forma inicial de salida del meta-modelo: {output.shape}")
                
                # Align shapes for loss calculation: both to (B, T_meta, H, W) if C=1
                # Squeeze the channel dimension (last dim) if it's 1 for both output and yb
                if output.dim() == 5 and output.shape[-1] == 1:
                    output = output.squeeze(-1) 

                if yb.dim() == 5 and yb.shape[-1] == 1:
                    yb = yb.squeeze(-1) 
                
                # Critical shape check
                if output.shape != yb.shape:
                    raise RuntimeError(f"Critical shape mismatch during meta-model training: output {output.shape}, yb {yb.shape}")
                
                loss = criterion(output, yb)
                loss.backward()
                optimizer.step()
                train_losses.append(loss.item())
                
            except RuntimeError as e:
                logger.error(f"Error en entrenamiento del meta-modelo: {e}")
                continue

        model.eval()
        val_losses = []
        with torch.no_grad():
            for Xv, yv in val_loader:
                try:
                    Xv, yv = Xv.to(DEVICE), yv.to(DEVICE)
                    
                    val_output = model(Xv)
                    
                    # Apply the same shape adjustments for validation
                    if val_output.dim() == 5 and val_output.shape[-1] == 1:
                        val_output = val_output.squeeze(-1)

                    if yv.dim() == 5 and yv.shape[-1] == 1:
                        yv = yv.squeeze(-1)

                    if val_output.shape != yv.shape:
                        raise RuntimeError(f"Critical shape mismatch during meta-model validation: val_output {val_output.shape}, yv {yv.shape}")
                    
                    val_loss = criterion(val_output, yv)
                    val_losses.append(val_loss.item())
                    
                except Exception as e:
                    logger.error(f"Error en validacin del meta-modelo: {e}")
                    continue

        train_loss_avg = np.mean(train_losses) if train_losses else float('inf')
        val_loss_avg = np.mean(val_losses) if val_losses else float('inf')
        history["train_loss"].append(train_loss_avg)
        history["val_loss"].append(val_loss_avg)
        logger.info(f"[Meta] Epoch {ep+1}/{max_epochs} - Train Loss: {train_loss_avg:.4f} - Val Loss: {val_loss_avg:.4f}")

        if val_loss_avg < best_val:
            best_val = val_loss_avg
            counter = 0
            torch.save(model.state_dict(), save_path)
        else:
            counter += 1
            if counter > patience:
                logger.info(f"Early stopping activado en meta-modelo en epoch {ep+1}")
                break

    try:
        model.load_state_dict(torch.load(save_path, weights_only=False))
    except Exception as e:
        logger.error(f"Error al cargar el mejor meta-modelo: {e}")
        
    return model, history

# %% [markdown]
# ## 6) Carga y preprocesamiento de datos

# %%
def load_and_preprocess_data():
    try:
        ds_full = xr.open_dataset(FULL_NC)
        logger.info(f"Dataset completo cargado: {FULL_NC}")
    except Exception as e:
        logger.error(f"No se pudo abrir FULL_NC: {e}")
        raise e

    try:
        ds_feat = xr.open_dataset(FEATURES_NC)
        logger.info(f"Dataset de features cargado: {FEATURES_NC}")
    except Exception as e:
        logger.error(f"No se pudo abrir FEATURES_NC: {e}")
        raise e

    try:
        gdf = gpd.read_file(SHP_PATH)
        if gdf.crs is None:
            logger.warning("Shapefile sin CRS definido. Asignando EPSG:4326 por defecto.")
            gdf = gdf.set_crs(epsg=4326, allow_override=True)
        elif gdf.crs.to_epsg() != 4326:
            gdf = gdf.to_crs(epsg=4326)
        logger.info(f"Shapefile cargado y CRS validado: {SHP_PATH}")
    except Exception as e:
        logger.error(f"No se pudo cargar o transformar shapefile: {e}")
        raise e

    times = ds_full.time.values.astype("datetime64[M]")
    REF = np.datetime64("2025-02", "M")

    indices_ref = np.where(times == REF)[0]
    if len(indices_ref) == 0:
        logger.error(f"Fecha de referencia {REF} no encontrada en el dataset de tiempos.")
        raise ValueError(f"Fecha {REF} no encontrada en los datos.")
    idx_ref = min(indices_ref[0], len(times) - 1)

    return ds_full, ds_feat, gdf, times, REF, idx_ref

ds_full, ds_feat, gdf, times, REF, idx_ref = load_and_preprocess_data()

dates_val = [str(REF - np.timedelta64(i, 'M')) for i in range(OUTPUT_HORIZON)][::-1]
dates_fc = [str(REF + np.timedelta64(i + 1, 'M')) for i in range(OUTPUT_HORIZON)]

# Definir representative_date_for_viz aqu para que est disponible globalmente
representative_date_for_viz = dates_val[0]


# %% [markdown]
# ## 7) Entrenamiento y evaluacin modelos base

# %%
def train_and_evaluate_base_models():
    METHODS = ["CEEMDAN", "TVFEMD", "FUSION"]
    BRANCHES = ["high", "medium", "low"]

    all_metrics = []
    # Key for preds_store/true_store now includes prediction start date and predicted month date
    # Value will be the path to the .npy file
    preds_store = {} # Key: (model_name, prediction_start_date_str, predicted_month_date_str) -> path_to_file.npy
    true_store = {}  # Key: (model_name_for_reference, prediction_start_date_str, predicted_month_date_str) -> path_to_file.npy
    histories = {}
    scalers_store = {} 

    T_full = ds_full.time.size
    ny, nx = ds_full.latitude.size, ds_full.longitude.size
    n_cells = ny * nx

    # Iterate over all possible starting points for predictions in the full dataset
    Nw_total = T_full - INPUT_WINDOW - OUTPUT_HORIZON + 1
    if Nw_total <= 0:
        logger.error("Not enough data to create any prediction windows.")
        return all_metrics, preds_store, true_store, histories, scalers_store

    # Pre-calculate all window details (dates) for efficiency
    all_window_start_dates = []
    all_predicted_dates_for_windows = []
    for i in range(Nw_total):
        # The window start date is the last month of the INPUT_WINDOW
        prediction_start_date_obj = pd.to_datetime(ds_full.time.values[i + INPUT_WINDOW -1]).to_period('M').to_timestamp()
        
        # The predicted months are from (INPUT_WINDOW_end + 1 month) to (INPUT_WINDOW_end + OUTPUT_HORIZON months)
        predicted_dates_obj = [pd.to_datetime(ds_full.time.values[i + INPUT_WINDOW + h_idx]).to_period('M').to_timestamp() for h_idx in range(OUTPUT_HORIZON)]

        all_window_start_dates.append(str(prediction_start_date_obj.date())[:-3]) # Format YYYY-MM
        all_predicted_dates_for_windows.append([str(d.date())[:-3] for d in predicted_dates_obj]) # Format YYYY-MM

    # Cache ys_raw for all windows, as it's common across all base models
    all_ys_raw_cached = np.stack([ds_full["total_precipitation"].values[i + INPUT_WINDOW : i + INPUT_WINDOW + OUTPUT_HORIZON].reshape(OUTPUT_HORIZON, n_cells) for i in range(Nw_total)], axis=0)


    for method in METHODS:
        for branch in BRANCHES:
            name = f"{method}_{branch}"
            logger.info(f"Processing {name}")

            # Get data for current method/branch
            Xarr_method_branch = ds_feat[name].values
            
            # Recreate Xs for this method/branch specifically for the full period
            Xs_method_branch_raw = np.stack([Xarr_method_branch[i : i + INPUT_WINDOW].reshape(INPUT_WINDOW, n_cells) for i in range(Nw_total)], axis=0)
            
            if branch == "low":
                months_all = pd.to_datetime(ds_full.time.values).month.values
                s_all = np.sin(2 * np.pi * months_all / 12)
                c_all = np.cos(2 * np.pi * months_all / 12)
                
                Ss_windowed_all = np.stack([s_all[i : i + INPUT_WINDOW] for i in range(Nw_total)], axis=0) 
                Cs_windowed_all = np.stack([c_all[i : i + INPUT_WINDOW] for i in range(Nw_total)], axis=0) 
                
                global_seasonal_features_all = np.stack([Ss_windowed_all, Cs_windowed_all], axis=-1) 
                
                Xs = np.concatenate([Xs_method_branch_raw, global_seasonal_features_all], axis=2)
                current_n_feats = n_cells + 2 
            else:
                Xs = Xs_method_branch_raw 
                current_n_feats = n_cells 
            
            logger.info(f"Shape of Xs for {name} before scaling: {Xs.shape}, calculated n_feats: {current_n_feats}")

            Xs_reshaped_for_scaler = Xs.reshape(-1, current_n_feats)
            ys_reshaped_for_scaler = all_ys_raw_cached.reshape(-1, n_cells) 

            scX = StandardScaler().fit(Xs_reshaped_for_scaler)
            scY = StandardScaler().fit(ys_reshaped_for_scaler)

            joblib.dump(scX, MODELS_OUTPUT / f"scaler_X_{name}.joblib")
            joblib.dump(scY, MODELS_OUTPUT / f"scaler_Y_{name}.joblib")
            scalers_store[name] = {"scX": scX, "scY": scY}
            logger.info(f"Scalers saved for {name}")

            Xs_s = scX.transform(Xs_reshaped_for_scaler).reshape(Xs.shape)
            ys_s = scY.transform(ys_reshaped_for_scaler).reshape(all_ys_raw_cached.shape)

            # Define split for base models training and validation based on the `REF` date
            train_windows_indices = []
            val_windows_indices = []
            
            first_val_month_str = dates_val[0] # e.g., '2024-12'
            
            for k in range(Nw_total):
                # If any of the predicted months for this window `k` fall into the validation period
                # (defined by `dates_val`), then this window belongs to validation for base models.
                overlaps_val = False
                for pred_date_str in all_predicted_dates_for_windows[k]:
                    if pred_date_str in dates_val:
                        overlaps_val = True
                        break
                
                if overlaps_val:
                    val_windows_indices.append(k)
                else:
                    train_windows_indices.append(k)

            X_tr, y_tr = Xs_s[train_windows_indices], ys_s[train_windows_indices]
            X_va, y_va = Xs_s[val_windows_indices], ys_s[val_windows_indices]

            logger.info(f"Base model split for {name}: Train samples: {len(X_tr)}, Val samples: {len(X_va)}")

            if len(X_tr) == 0:
                logger.warning(f"No training samples for base model {name}. Skipping training.")
                continue 

            train_ds = BaseDataset(X_tr, y_tr)
            val_ds = BaseDataset(X_va, y_va)
            tr_ld = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True)
            va_ld = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False)

            if branch == "low":
                model = GRUEncoderDecoder(current_n_feats, 128, OUTPUT_HORIZON, n_cells, dropout=0.1) # Added dropout
            else:
                model = TransformerModel(current_n_feats, OUTPUT_HORIZON, n_cells, dropout=0.1) # Added dropout
            model.to(DEVICE)
            opt = optim.Adam(model.parameters(), lr=LR)
            crit = nn.L1Loss() 

            save_path = MODELS_OUTPUT / f"best_base_{name}.pt"
            model, hist = train_base_model(model, opt, crit, tr_ld, va_ld, save_path=save_path)
            histories[name] = hist
            plot_training_history(hist, title=f"Prdida Entrenamiento {name}")

            model.eval()
            with torch.no_grad():
                # Predict on ALL historical windows (Xs_s) to populate preds_store and true_store for meta-model
                all_preds_scaled = model(torch.tensor(Xs_s, dtype=torch.float32, device=DEVICE)).cpu().numpy()
            
            # Store predictions and true values for ALL windows for meta-model on disk
            for window_idx in range(Nw_total):
                prediction_start_date_str = all_window_start_dates[window_idx]
                
                # Create a specific directory for this model and prediction start date
                model_pred_dir = PREDS_DIR / name / prediction_start_date_str
                model_pred_dir.mkdir(parents=True, exist_ok=True)
                
                for h_idx in range(OUTPUT_HORIZON):
                    predicted_month_date_str = all_predicted_dates_for_windows[window_idx][h_idx]
                    
                    pred_map_inv = scY.inverse_transform(all_preds_scaled[window_idx, h_idx, :].reshape(1, n_cells)).reshape(ny, nx)
                    true_map_inv = scY.inverse_transform(ys_s[window_idx, h_idx, :].reshape(1, n_cells)).reshape(ny, nx)
                    
                    # Store PATHS in preds_store/true_store
                    # Key is (model_name, prediction_start_date, predicted_month_date)
                    pred_file_path = model_pred_dir / f"pred_{predicted_month_date_str}_h{h_idx+1}.npy"
                    true_file_path = model_pred_dir / f"true_{predicted_month_date_str}_h{h_idx+1}.npy"
                    
                    np.save(pred_file_path, pred_map_inv)
                    np.save(true_file_path, true_map_inv)
                    
                    preds_store[(name, prediction_start_date_str, predicted_month_date_str)] = str(pred_file_path)
                    # For true_store, we can use a generic key as true values are the same across models
                    true_store[(f"{METHODS[0]}_{BRANCHES[0]}", prediction_start_date_str, predicted_month_date_str)] = str(true_file_path)
                    
                    # Also collect metrics for the validation set (if this window is in validation)
                    if window_idx in val_windows_indices:
                        rmse, mae, mape, r2 = evaluate_metrics_np(true_map_inv.ravel(), pred_map_inv.ravel())
                        all_metrics.append({
                            "model": name, "branch": branch, "horizon": h_idx+1,
                            "type": "validation", "date": predicted_month_date_str, # Date is the predicted month
                            "RMSE": rmse, "MAE": mae, "MAPE": mape, "R2": r2
                        })

    return all_metrics, preds_store, true_store, histories, scalers_store

all_metrics, preds_store, true_store, histories, scalers_store = train_and_evaluate_base_models()

# %% [markdown]
# ## 8) Preparar meta-dataset y entrenar meta-modelo

# %%
def prepare_meta_dataset(preds_store, true_store, dates_val, METHODS, BRANCHES, ny, nx, output_horizon):
    Xm_paths, ym_paths = [], [] # Store paths to the .npy files

    T_full = ds_full.time.size
    Nw_total = T_full - INPUT_WINDOW - output_horizon + 1
    
    # Recalculate window dates as they are needed to reconstruct keys for preds_store
    all_window_start_dates = []
    all_predicted_dates_for_windows = []
    for i in range(Nw_total):
        prediction_start_date_obj = pd.to_datetime(ds_full.time.values[i + INPUT_WINDOW -1]).to_period('M').to_timestamp()
        predicted_dates_obj = [pd.to_datetime(ds_full.time.values[i + INPUT_WINDOW + h_idx]).to_period('M').to_timestamp() for h_idx in range(output_horizon)]
        all_window_start_dates.append(str(prediction_start_date_obj.date())[:-3])
        all_predicted_dates_for_windows.append([str(d.date())[:-3] for d in predicted_dates_obj])

    # Define the meta-model's training and validation period
    # Meta-model train: all windows where the predicted period ends BEFORE dates_val[0]
    # Meta-model val: all windows where the predicted period starts in dates_val
    
    meta_train_indices = []
    meta_val_indices = []
            
    for k in range(Nw_total):
        # Check if any of the predicted months for this window `k` fall into the BASE MODEL validation period
        overlaps_base_val = False
        for pred_date_str in all_predicted_dates_for_windows[k]:
            if pred_date_str in dates_val:
                overlaps_base_val = True
                break
        
        if overlaps_base_val:
            meta_val_indices.append(k)
        else:
            meta_train_indices.append(k)
    
    logger.info(f"Meta-model raw samples (all possible windows): {Nw_total}")
    logger.info(f"Meta-model training samples (from base model train period): {len(meta_train_indices)}")
    logger.info(f"Meta-model validation samples (from base model val period): {len(meta_val_indices)}")
    
    # Populate Xm_paths and ym_paths based on these indices
    # Each Xm_path will be a list of lists of paths: [horizon_idx][model_idx] -> path
    # Each ym_path will be a list of paths: [horizon_idx] -> path
    for window_idx in range(Nw_total): # Iterate over all possible windows
        current_window_start_date = all_window_start_dates[window_idx]
        current_predicted_dates = all_predicted_dates_for_windows[window_idx]
        
        X_sample_paths_for_this_window = [] 
        y_sample_paths_for_this_window = [] 

        full_sample_complete = True
        for h_idx in range(output_horizon):
            predicted_month_date_str = current_predicted_dates[h_idx]
            
            preds_for_horizon_paths = []
            horizon_complete = True
            for br in BRANCHES:
                for me in METHODS:
                    key = (f"{me}_{br}", current_window_start_date, predicted_month_date_str)
                    if key in preds_store:
                        preds_for_horizon_paths.append(preds_store[key]) # Store the path
                    else:
                        logger.warning(f"Prediction path not found for meta-model input: {key}. Skipping sample window {window_idx}.")
                        horizon_complete = False
                        break
                if not horizon_complete: break
            
            if horizon_complete:
                X_sample_paths_for_this_window.append(preds_for_horizon_paths) # Add list of model paths for this horizon
                
                # Use a generic base model key for true_store lookup
                true_key = (f"{METHODS[0]}_{BRANCHES[0]}", current_window_start_date, predicted_month_date_str)
                if true_key in true_store:
                    y_sample_paths_for_this_window.append(true_store[true_key]) # Store the path
                else:
                    logger.warning(f"True value path not found for meta-model target: {true_key}. Skipping sample window {window_idx}.")
                    full_sample_complete = False
                    break
            else:
                full_sample_complete = False
                break
        
        if full_sample_complete and len(X_sample_paths_for_this_window) == output_horizon:
            Xm_paths.append(X_sample_paths_for_this_window)
            ym_paths.append(y_sample_paths_for_this_window)

    if not Xm_paths:
        logger.error("No complete samples for meta-dataset after verification. Check data availability and slicing.")
        return None, None, None, None # Return None for all splits

    # Split the paths into train and validation based on original indices
    # This assumes that the order of samples in Xm_paths/ym_paths matches the original window_idx order.
    # It should, as they are appended in the same `for window_idx in range(Nw_total)` loop.
    
    # We must filter Xm_paths, ym_paths using meta_train_indices and meta_val_indices directly
    # as these indices refer to the `Nw_total` windows, which are the same as `Xm_paths` indices.
    
    # Need to convert list of lists of paths to numpy object array first for proper indexing
    Xm_paths_all_np = np.array(Xm_paths, dtype=object)
    ym_paths_all_np = np.array(ym_paths, dtype=object)

    Xm_tr_paths = Xm_paths_all_np[meta_train_indices]
    ym_tr_paths = ym_paths_all_np[meta_train_indices]
    Xm_va_paths = Xm_paths_all_np[meta_val_indices]
    ym_va_paths = ym_paths_all_np[meta_val_indices]

    # Handle cases where training/validation sets might be empty after split
    if len(Xm_tr_paths) == 0 and len(Xm_va_paths) > 0:
        logger.warning("Meta-model training set is empty. Using all validation data for training.")
        Xm_tr_paths, ym_tr_paths = Xm_va_paths, ym_va_paths
        Xm_va_paths, ym_va_paths = np.array([]), np.array([]) # Empty validation set
    elif len(Xm_tr_paths) == 0 and len(Xm_va_paths) == 0:
        logger.error("No training or validation samples for meta-model after split. Cannot proceed with training.")
        return None, None, None, None # Indicate untrainable
    elif len(Xm_va_paths) == 0 and len(Xm_tr_paths) > 0:
        # If no explicit validation set, create a small one from training data
        if len(Xm_tr_paths) > 1: # Need at least 2 samples to split
            train_idx_split, val_idx_split = train_test_split(np.arange(len(Xm_tr_paths)), test_size=max(1, int(len(Xm_tr_paths)*0.1)), random_state=42)
            Xm_va_paths_new, ym_va_paths_new = Xm_tr_paths[val_idx_split], ym_tr_paths[val_idx_split]
            Xm_tr_paths_new, ym_tr_paths_new = Xm_tr_paths[train_idx_split], ym_tr_paths[train_idx_split]
            logger.warning(f"Meta-model validation set was empty. Created small validation set from training data: {len(Xm_va_paths_new)} samples.")
            Xm_tr_paths, ym_tr_paths = Xm_tr_paths_new, ym_tr_paths_new
            Xm_va_paths, ym_va_paths = Xm_va_paths_new, ym_va_paths_new
        else:
            logger.warning("Meta-model has too few samples for proper train/val split. Training on all available data with no separate validation.")
            # Keep Xm_tr_paths, ym_tr_paths as they are
            Xm_va_paths, ym_va_paths = np.array([]), np.array([]) # Explicitly empty val set


    return Xm_tr_paths, ym_tr_paths, Xm_va_paths, ym_va_paths


ny, nx = ds_full.latitude.size, ds_full.longitude.size
METHODS = ["CEEMDAN", "TVFEMD", "FUSION"]
BRANCHES = ["high", "medium", "low"]

Xm_tr_paths, ym_tr_paths, Xm_va_paths, ym_va_paths = prepare_meta_dataset(preds_store, true_store, dates_val, METHODS, BRANCHES, ny, nx, OUTPUT_HORIZON)

if Xm_tr_paths is None or len(Xm_tr_paths) == 0: 
    logger.error("Meta-model dataset could not be prepared or has insufficient training samples. Skipping meta-model training and evaluation.")
    meta_model_trainable = False
else:
    meta_model_trainable = True
    met_tr = MetaDataset(Xm_tr_paths, ym_tr_paths)
    met_va = MetaDataset(Xm_va_paths, ym_va_paths)
    
    # Increase num_workers for faster data loading from disk
    md_tr = DataLoader(met_tr, batch_size=BATCH_SIZE, shuffle=True, num_workers=min(os.cpu_count(), 4)) 
    md_va = DataLoader(met_va, batch_size=BATCH_SIZE, num_workers=min(os.cpu_count(), 4)) 

    meta_model = UNetConvLSTMMeta(num_models=len(METHODS)*len(BRANCHES), ny=ny, nx=nx, output_horizon=OUTPUT_HORIZON, channels=1).to(DEVICE)
    opt_meta = optim.Adam(meta_model.parameters(), lr=LR)
    crit_meta = nn.L1Loss() 

    save_path_meta = MODELS_OUTPUT / "best_meta_model.pt"
    meta_model, hist_meta = train_meta_model(meta_model, opt_meta, crit_meta, md_tr, md_va, save_path=save_path_meta)
    plot_training_history(hist_meta, title="Prdida Entrenamiento Meta-modelo")

    torch.save(meta_model.state_dict(), MODELS_OUTPUT / "meta_model_final.pt")
    logger.info("Meta-modelo entrenado y guardado.")

    meta_model.eval()
    yhat_full = np.array([])
    ym_va_full = np.array([])

    if len(md_va) > 0: # Only predict if there's actual validation data
        with torch.no_grad():
            yhat_list = []
            ym_va_list = []
            for Xv_batch, yv_batch in md_va:
                Xv_batch_cuda = Xv_batch.to(DEVICE)
                val_output = meta_model(Xv_batch_cuda).cpu().numpy()
                
                # Adjust shapes before appending to lists if needed
                if val_output.ndim == 5 and val_output.shape[-1] == 1:
                    val_output = val_output.squeeze(-1) # (B, T_meta, H, W)
                if yv_batch.ndim == 5 and yv_batch.shape[-1] == 1:
                    yv_batch = yv_batch.squeeze(-1).cpu().numpy() # (B, T_meta, H, W)

                yhat_list.append(val_output)
                ym_va_list.append(yv_batch)
            
            yhat_full = np.concatenate(yhat_list, axis=0)
            ym_va_full = np.concatenate(ym_va_list, axis=0)
    else: 
        logger.warning("No meta-model validation samples to predict on for full evaluation. Metrics will be NaN.")
    

    # Calculate metrics for Meta-Model for each horizon
    meta_metrics_list_per_horizon = []
    if len(ym_va_full) > 0: # Only calculate if there's actual validation data
        for h in range(OUTPUT_HORIZON):
            rmse_m_h, mae_m_h, mape_m_h, r2_m_h = evaluate_metrics_np(ym_va_full[:, h, :, :].ravel(), yhat_full[:, h, :, :].ravel())
            meta_metrics_list_per_horizon.append({
                "model": "Meta-Modelo", 
                "branch": "Meta", 
                "horizon": h+1, 
                "RMSE": rmse_m_h, 
                "MAE": mae_m_h, 
                "MAPE": mape_m_h, 
                "R2": r2_m_h
            })
    else:
        logger.warning("No meta-model validation samples for metric calculation. Meta-model metrics will be NaN.")
        for h in range(OUTPUT_HORIZON): # Still add rows with NaNs to table
            meta_metrics_list_per_horizon.append({
                "model": "Meta-Modelo", "branch": "Meta", "horizon": h+1, 
                "RMSE": np.nan, "MAE": np.nan, "MAPE": np.nan, "R2": np.nan
            })


    df_meta = pd.DataFrame(meta_metrics_list_per_horizon)
    csv_meta_path = MODELS_OUTPUT / f"metrics_meta_ref{str(REF)}.csv"
    df_meta.to_csv(csv_meta_path, index=False)
    logger.info(f"Mtricas meta-modelo guardadas en {csv_meta_path}")

# **INITIALIZE AND POPULATE meta_preds AND meta_true FOR VISUALIZATION**
meta_preds = {} 
meta_true = {}  

# Only populate if meta_model was trained successfully and has predictions
if meta_model_trainable and len(yhat_full) > 0:
    # Use only the first sample from yhat_full and ym_va_full for visualization of all horizons
    if yhat_full.shape[0] > 0:
        for h in range(OUTPUT_HORIZON):
            date_for_viz = dates_val[h] 
            meta_preds[(f"Meta", date_for_viz)] = yhat_full[0, h, :, :] 
            meta_true[(f"Meta", date_for_viz)] = ym_va_full[0, h, :, :] 
        logger.info("Meta-model visualization data populated from first validation sample.")
    else:
        logger.warning("No validation samples available for meta-model visualization.")


# **CALCULATE AND STORE METRICS FOR "ENSEMBLE AVERAGE"**
ensemble_avg_preds = {} 
ensemble_avg_true = {}  
ensemble_metrics_list = [] 

# Iterate through all unique predicted target dates
unique_predicted_target_dates = sorted(list(set([k[2] for k in preds_store.keys()])))

for target_date_str in unique_predicted_target_dates:
    for h in range(1, OUTPUT_HORIZON + 1): 
        current_preds_for_ensemble = []
        current_true_for_ensemble = [] 

        for method in METHODS:
            for branch in BRANCHES:
                target_date_obj = pd.Period(target_date_str).to_timestamp()
                pred_start_date_obj = target_date_obj - pd.DateOffset(months=h-1)
                pred_start_date_str = str(pred_start_date_obj.date())[:-3]
                
                key = (f"{method}_{branch}", pred_start_date_str, target_date_str)
                
                if key in preds_store and key in true_store:
                    pred_map = np.load(preds_store[key])
                    true_map = np.load(true_store[key])
                    current_preds_for_ensemble.append(pred_map)
                    current_true_for_ensemble.append(true_map)
        
        if current_preds_for_ensemble:
            avg_map = np.mean(np.stack(current_preds_for_ensemble), axis=0) # (H, W)
            # Pick one true map, as they should be identical
            true_map_for_ensemble = current_true_for_ensemble[0] 

            ensemble_avg_preds[(f"Ensemble_Average", target_date_str, h)] = avg_map
            ensemble_avg_true[(f"Ensemble_Average", target_date_str, h)] = true_map_for_ensemble
            
            # If this target_date is part of `dates_val`, add to all_metrics for comparative table
            if target_date_str in dates_val:
                rmse_e, mae_e, mape_e, r2_e = evaluate_metrics_np(true_map_for_ensemble.ravel(), avg_map.ravel())
                ensemble_metrics_list.append({
                    "model": "Ensemble_Average", 
                    "branch": "Avg", 
                    "horizon": h, 
                    "type": "validation", 
                    "date": target_date_str,
                    "RMSE": rmse_e, "MAE": mae_e, "MAPE": mape_e, "R2": r2_e
                })

# Add ensemble metrics to all_metrics
all_metrics.extend(ensemble_metrics_list)


# %% [markdown]
# ## 9) Visualizacin

# %%
lat = ds_full.latitude.values
lon = ds_full.longitude.values
boyaca_gdf = gpd.read_file(SHP_PATH)
if boyaca_gdf.crs is None:
    boyaca_gdf = boyaca_gdf.set_crs(epsg=4326, allow_override=True)
elif boyaca_gdf.crs.to_epsg() != 4326:
    boyaca_gdf = boyaca_gdf.to_crs(epsg=4326)

for date_idx, date in enumerate(dates_val): # Iterate over each validation date (e.g., 2024-12, 2025-01, 2025-02)
    vmin = None; vmax = None
    all_vals = []
    
    # Determine the correct prediction_start_date for this specific `date` and `date_idx` (representing horizon step)
    date_obj = pd.Period(date).to_timestamp()
    pred_start_date_obj_for_this_val_run = date_obj - pd.DateOffset(months=date_idx)
    pred_start_date_str_for_this_val_run = str(pred_start_date_obj_for_this_val_run.date())[:-3]

    # Base models: predictions stored with (model_name, prediction_start_date_str, predicted_month_date_str)
    for method in METHODS:
        for branch in BRANCHES:
            key = (f"{method}_{branch}", pred_start_date_str_for_this_val_run, date) 
            if key in preds_store:
                all_vals.append(np.load(preds_store[key]).ravel()) # Load from path
            else:
                logger.warning(f"No prediction found for {key} in preds_store for visualization.")

    # Meta-Model: Only visualize if trained and if there's a sample for this specific horizon
    if meta_model_trainable:
        meta_key = (f"Meta", date) # Key for meta_preds, which stores data per horizon.
        if meta_key in meta_preds:
            all_vals.append(meta_preds[meta_key].ravel())
        else:
            logger.warning(f"No meta-model prediction found for {meta_key} for visualization.")

    # Ensemble Average:
    ensemble_key = (f"Ensemble_Average", date, date_idx + 1) # Key includes specific horizon
    if ensemble_key in ensemble_avg_preds:
        all_vals.append(ensemble_avg_preds[ensemble_key].ravel())
    else:
        logger.warning(f"No ensemble average prediction found for {ensemble_key} for visualization.")


    if all_vals:
        all_vals = np.concatenate(all_vals)
        vmin = np.nanmin(all_vals) 
        vmax = np.nanmax(all_vals) 
        if vmin < 0: vmin = 0 
        if vmax == 0: vmax = 1 

    fig, axs = plt.subplots(3, 3, figsize=(15, 12), subplot_kw={"projection": ccrs.PlateCarree()})
    fig.suptitle(f"Prediccin Precipitacin para {date} (Horizonte {date_idx+1})", fontsize=16)
    pcm = None # Initialize pcm before the loop
    for i, branch in enumerate(BRANCHES):
        for j, method in enumerate(METHODS):
            ax = axs[i, j]
            ax.set_extent(boyaca_gdf.total_bounds[[0, 2, 1, 3]], crs=ccrs.PlateCarree())
            ax.add_feature(cfeature.COASTLINE, linewidth=0.5)
            ax.add_feature(cfeature.BORDERS, linewidth=0.5)
            ax.add_geometries(boyaca_gdf.geometry, ccrs.PlateCarree(), edgecolor='black', facecolor='none')
            key = (f"{method}_{branch}", pred_start_date_str_for_this_val_run, date) # Reconstruct key for base model
            if key in preds_store:
                pcm = ax.pcolormesh(lon, lat, np.load(preds_store[key]), vmin=vmin, vmax=vmax, 
                                 cmap="Blues", shading='auto', alpha=0.8)
            ax.set_title(f"{method} - {branch}")
    if pcm is not None: # Only draw colorbar if pcm was actually created
        fig.colorbar(pcm, ax=axs, orientation='horizontal', fraction=0.05, pad=0.05, 
                   label='Precipitacin (mm)', extend='both')
    plt.show()

    fig, axs = plt.subplots(3, 3, figsize=(15, 12), subplot_kw={"projection": ccrs.PlateCarree()})
    fig.suptitle(f"MAPE Precipitacin para {date} (Horizonte {date_idx+1})", fontsize=16)
    pcm_mape = None # Initialize pcm_mape
    for i, branch in enumerate(BRANCHES):
        for j, method in enumerate(METHODS):
            ax = axs[i, j]
            ax.set_extent(boyaca_gdf.total_bounds[[0, 2, 1, 3]], crs=ccrs.PlateCarree())
            ax.add_feature(cfeature.COASTLINE, linewidth=0.5)
            ax.add_feature(cfeature.BORDERS, linewidth=0.5)
            ax.add_geometries(boyaca_gdf.geometry, ccrs.PlateCarree(), edgecolor='black', facecolor='none')
            key = (f"{method}_{branch}", pred_start_date_str_for_this_val_run, date) # Reconstruct key for base model
            if key in preds_store and key in true_store:
                pred_map = np.load(preds_store[key])
                true_map = np.load(true_store[key])
                mape_arr = np.abs((true_map - pred_map) / (true_map + 1e-5)) * 100
                mape_arr = np.clip(mape_arr, 0, 200) 
                pcm_mape = ax.pcolormesh(lon, lat, mape_arr, vmin=0, vmax=200, 
                                  cmap="Reds", shading='auto', alpha=0.8)
            ax.set_title(f"{method} - {branch}")
    if pcm_mape is not None: # Only draw colorbar if pcm_mape was created
        fig.colorbar(pcm_mape, ax=axs, orientation='horizontal', fraction=0.05, pad=0.05, 
                   label='MAPE (%)', extend='max')
    plt.show()

# %% [markdown]
# ### 9.1) Visualizacin del Meta-modelo y Ensemble Average por Horizonte

# %%
if meta_model_trainable and len(dates_val) > 0:
    for h_idx in range(OUTPUT_HORIZON):
        current_date_for_horizon_viz = dates_val[h_idx] # The specific date for this horizon
        
        all_combined_vals_h = []
        meta_key_h = (f"Meta", current_date_for_horizon_viz)
        if meta_key_h in meta_preds:
            all_combined_vals_h.append(meta_preds[meta_key_h].ravel())
        
        ensemble_key_h = (f"Ensemble_Average", current_date_for_horizon_viz, h_idx + 1)
        if ensemble_key_h in ensemble_avg_preds:
            all_combined_vals_h.append(ensemble_avg_preds[ensemble_key_h].ravel())

        vmin_combined_h = None
        vmax_combined_h = None
        if all_combined_vals_h:
            all_combined_vals_h = np.concatenate(all_combined_vals_h)
            vmin_combined_h = np.nanmin(all_combined_vals_h)
            vmax_combined_h = np.nanmax(all_combined_vals_h)
            if vmin_combined_h < 0: vmin_combined_h = 0
            if vmax_combined_h == 0: vmax_combined_h = 1 


        # Visualize prediction maps for Meta-Model and Ensemble Average for the current horizon
        fig, axs = plt.subplots(1, 2, figsize=(20, 10), subplot_kw={"projection": ccrs.PlateCarree()})
        fig.suptitle(f"Predicciones Combinadas para {current_date_for_horizon_viz} (Horizonte {h_idx+1})", fontsize=16)

        # Meta-Model Prediction
        ax_meta = axs[0]
        ax_meta.set_extent(boyaca_gdf.total_bounds[[0, 2, 1, 3]], crs=ccrs.PlateCarree())
        ax_meta.add_feature(cfeature.COASTLINE, linewidth=0.5)
        ax_meta.add_feature(cfeature.BORDERS, linewidth=0.5)
        ax_meta.add_geometries(boyaca_gdf.geometry, ccrs.PlateCarree(), edgecolor='black', facecolor='none')
        pcm_meta = None # Initialize
        if meta_key_h in meta_preds:
            pcm_meta = ax_meta.pcolormesh(lon, lat, meta_preds[meta_key_h], vmin=vmin_combined_h, vmax=vmax_combined_h, 
                                          cmap="Blues", shading='auto', alpha=0.8)
        ax_meta.set_title("Meta-Modelo")

        # Ensemble Average Prediction
        ax_ensemble = axs[1]
        ax_ensemble.set_extent(boyaca_gdf.total_bounds[[0, 2, 1, 3]], crs=ccrs.PlateCarree())
        ax_ensemble.add_feature(cfeature.COASTLINE, linewidth=0.5)
        ax_ensemble.add_feature(cfeature.BORDERS, linewidth=0.5)
        ax_ensemble.add_geometries(boyaca_gdf.geometry, ccrs.PlateCarree(), edgecolor='black', facecolor='none')
        pcm_ensemble = None # Initialize
        if ensemble_key_h in ensemble_avg_preds:
            pcm_ensemble = ax_ensemble.pcolormesh(lon, lat, ensemble_avg_preds[ensemble_key_h], vmin=vmin_combined_h, vmax=vmax_combined_h, 
                                                  cmap="Blues", shading='auto', alpha=0.8)
        ax_ensemble.set_title("Ensemble Average")

        if pcm_meta is not None: # Use pcm_meta for colorbar as vmin/vmax are shared
            fig.colorbar(pcm_meta, ax=axs.ravel().tolist(), orientation='horizontal', fraction=0.05, pad=0.05, 
                         label='Precipitacin (mm)', extend='both')
        plt.show()

        # Visualize MAPE error maps for Meta-Model and Ensemble Average for the current horizon
        fig, axs = plt.subplots(1, 2, figsize=(20, 10), subplot_kw={"projection": ccrs.PlateCarree()})
        fig.suptitle(f"MAPE Combinado para {current_date_for_horizon_viz} (Horizonte {h_idx+1})", fontsize=16)

        # Meta-Model MAPE
        ax_meta_mape = axs[0]
        ax_meta_mape.set_extent(boyaca_gdf.total_bounds[[0, 2, 1, 3]], crs=ccrs.PlateCarree())
        ax_meta_mape.add_feature(cfeature.COASTLINE, linewidth=0.5)
        ax_meta_mape.add_feature(cfeature.BORDERS, linewidth=0.5)
        ax_meta_mape.add_geometries(boyaca_gdf.geometry, ccrs.PlateCarree(), edgecolor='black', facecolor='none')
        pcm_meta_mape = None # Initialize
        if meta_key_h in meta_preds and meta_key_h in meta_true:
            mape_arr_meta = np.abs((meta_true[meta_key_h] - meta_preds[meta_key_h]) / (meta_true[meta_key_h] + 1e-5)) * 100
            mape_arr_meta = np.clip(mape_arr_meta, 0, 200)
            pcm_meta_mape = ax_meta_mape.pcolormesh(lon, lat, mape_arr_meta, vmin=0, vmax=200, 
                                                    cmap="Reds", shading='auto', alpha=0.8)
        ax_meta_mape.set_title("Meta-Modelo MAPE")

        # Ensemble Average MAPE
        ax_ensemble_mape = axs[1]
        ax_ensemble_mape.set_extent(boyaca_gdf.total_bounds[[0, 2, 1, 3]], crs=ccrs.PlateCarree())
        ax_ensemble_mape.add_feature(cfeature.COASTLINE, linewidth=0.5)
        ax_ensemble_mape.add_feature(cfeature.BORDERS, linewidth=0.5)
        ax_ensemble_mape.add_geometries(boyaca_gdf.geometry, ccrs.PlateCarree(), edgecolor='black', facecolor='none')
        pcm_ensemble_mape = None # Initialize
        if ensemble_key_h in ensemble_avg_preds and ensemble_key_h in ensemble_avg_true:
            mape_arr_ensemble = np.abs((ensemble_avg_true[ensemble_key_h] - ensemble_avg_preds[ensemble_key_h]) / (ensemble_avg_true[ensemble_key_h] + 1e-5)) * 100
            mape_arr_ensemble = np.clip(mape_arr_ensemble, 0, 200)
            pcm_ensemble_mape = ax_ensemble_mape.pcolormesh(lon, lat, mape_arr_ensemble, vmin=0, vmax=200, 
                                                            cmap="Reds", shading='auto', alpha=0.8)
        ax_ensemble_mape.set_title("Ensemble Average MAPE")

        if pcm_meta_mape is not None: # Use pcm_meta_mape for colorbar
            fig.colorbar(pcm_meta_mape, ax=axs.ravel().tolist(), orientation='horizontal', fraction=0.05, pad=0.05, 
                         label='MAPE (%)', extend='max')
        plt.show()
else:
    logger.info("Skipping meta-model and ensemble average visualization as no trainable samples were found.")


# %% [markdown]
# ### 9.2) Tabla Comparativa de Mtricas por Modelo y Fecha

# %%
# Calculate additional metrics
def calculate_additional_metrics(y_true, y_pred):
    """Calculates additional metrics between true and predicted values."""
    # Ensure no NaNs are passed into calculations
    valid_indices = ~np.isnan(y_true) & ~np.isnan(y_pred)
    y_true_clean = y_true[valid_indices]
    y_pred_clean = y_pred[valid_indices]

    if y_true_clean.size == 0:
        return {"NSE": np.nan, "Rel_Bias(%)": np.nan, "NMAE(%)": np.nan, "Pearson_Corr": np.nan}

    # Nash-Sutcliffe Efficiency (NSE)
    mean_obs = np.mean(y_true_clean)
    y_true_variance = np.sum((y_true_clean - mean_obs)**2)
    if y_true_variance < 1e-6: # Using a small threshold for "constant" true values
        nse = np.nan
    else:
        nse = 1 - (np.sum((y_true_clean - y_pred_clean)**2) / y_true_variance)
    
    # Relative Bias (Bias)
    if np.sum(y_true_clean) == 0:
        rel_bias = np.nan
    else:
        rel_bias = np.sum(y_pred_clean - y_true_clean) / np.sum(y_true_clean) * 100
    
    # Normalized Mean Absolute Error (NMAE)
    if np.mean(y_true_clean) == 0:
        nmae = np.nan
    else:
        nmae = np.mean(np.abs(y_true_clean - y_pred_clean)) / np.mean(y_true_clean) * 100
    
    # Pearson Correlation
    if np.std(y_true_clean) > 0 and np.std(y_pred_clean) > 0:
        pearson_corr = np.corrcoef(y_true_clean, y_pred_clean)[0, 1]
    else:
        pearson_corr = np.nan
    
    return {
        "NSE": nse, 
        "Rel_Bias(%)": rel_bias,
        "NMAE(%)": nmae,
        "Pearson_Corr": pearson_corr
    }

# Prepare data for the comparative table
comparative_metrics = []

# Process metrics for base models and Ensemble_Average (already added to all_metrics)
for metric in all_metrics:
    if metric["type"] == "validation":  # Only include validation data
        data_row = {
            "Modelo": metric["model"],
            "Rama": metric["branch"],
            "Fecha": metric["date"],
            "Horizonte": metric["horizon"],
            "RMSE": metric["RMSE"],
            "MAE": metric["MAE"],
            "MAPE(%)": metric["MAPE"],
            "R": metric["R2"]
        }
        
        model_name = metric["model"]
        date = metric["date"]
        horizon_val = metric["horizon"] # Get horizon from metric

        # Determine the correct store for predictions and true values
        if model_name == "Ensemble_Average":
            key = (model_name, date, horizon_val) 
            pred_data = ensemble_avg_preds.get(key)
            true_data = ensemble_avg_true.get(key)
        else: # For base models
            date_obj = pd.Period(date).to_timestamp()
            pred_start_date_obj = date_obj - pd.DateOffset(months=horizon_val-1)
            pred_start_date_str = str(pred_start_date_obj.date())[:-3]

            key_preds_store = (model_name, pred_start_date_str, date)
            
            pred_data_path = preds_store.get(key_preds_store)
            true_data_path = true_store.get(key_preds_store)
            
            pred_data = np.load(pred_data_path) if pred_data_path else None
            true_data = np.load(true_data_path) if true_data_path else None

        
        # Add additional metrics if we have observed and predicted data
        if pred_data is not None and true_data is not None:
            y_true_flat = true_data.flatten()
            y_pred_flat = pred_data.flatten()
            
            data_row["Precip_Media(mm)"] = np.mean(y_pred_flat)
            
            add_metrics = calculate_additional_metrics(y_true_flat, y_pred_flat)
            data_row.update(add_metrics)
        else:
            data_row["Precip_Media(mm)"] = np.nan 
            data_row.update({"NSE": np.nan, "Rel_Bias(%)": np.nan, "NMAE(%)": np.nan, "Pearson_Corr": np.nan})

        comparative_metrics.append(data_row)

# Add meta-model metrics (from df_meta, which now has metrics per horizon)
if meta_model_trainable and not df_meta.empty:
    for _, meta_row in df_meta.iterrows():
        data_row = {
            "Modelo": "Meta-Modelo",
            "Rama": "Meta",
            "Fecha": dates_val[int(meta_row["horizon"])-1], # Map horizon index back to date_val
            "Horizonte": meta_row["horizon"],
            "RMSE": meta_row["RMSE"],
            "MAE": meta_row["MAE"],
            "MAPE(%)": meta_row["MAPE"],
            "R": meta_row["R2"],
        }
        
        # Calculate additional metrics for the meta-model
        # Need to extract the relevant slice from yhat_full and ym_va_full
        if len(yhat_full) > 0 and len(ym_va_full) > 0: # Check if validation data exists for meta-model
            h_idx = int(meta_row["horizon"]) - 1 # Get 0-indexed horizon
            y_true_meta_h = ym_va_full[:, h_idx, :, :].ravel()
            y_pred_meta_h = yhat_full[:, h_idx, :, :].ravel()
            
            data_row["Precip_Media(mm)"] = np.mean(y_pred_meta_h)
            add_metrics_meta = calculate_additional_metrics(y_true_meta_h, y_pred_meta_h)
            data_row.update(add_metrics_meta)
        else:
            data_row["Precip_Media(mm)"] = np.nan
            data_row.update({"NSE": np.nan, "Rel_Bias(%)": np.nan, "NMAE(%)": np.nan, "Pearson_Corr": np.nan})

        comparative_metrics.append(data_row)


# Create DataFrame and display table
df_comparative = pd.DataFrame(comparative_metrics)

# Sort by Fecha, then by Modelo, then by Horizonte
df_comparative = df_comparative.sort_values(by=["Fecha", "Modelo", "Horizonte"]).reset_index(drop=True)

# Format numerical values for better visualization
format_cols = {
    "RMSE": "{:.3f}",
    "MAE": "{:.3f}", 
    "MAPE(%)": "{:.2f}",
    "R": "{:.3f}",
    "Precip_Media(mm)": "{:.2f}",
    "NSE": "{:.3f}",
    "Rel_Bias(%)": "{:.2f}",
    "NMAE(%)": "{:.2f}",
    "Pearson_Corr": "{:.3f}"
}

for col, fmt in format_cols.items():
    if col in df_comparative.columns:
        df_comparative[col] = pd.to_numeric(df_comparative[col], errors='coerce')
        df_comparative[col] = df_comparative[col].apply(lambda x: fmt.format(x) if not pd.isna(x) else "")

from IPython.display import display
display(df_comparative.style.set_caption("Tabla Comparativa de Mtricas por Modelo y Fecha"))

csv_comparative_path = MODELS_OUTPUT / f"comparative_metrics_ref{str(REF)}.csv"
df_comparative.to_csv(csv_comparative_path, index=False)
logger.info(f"Comparative metrics table saved to {csv_comparative_path}")

# %% [markdown]
# ### 9.3) Resumen Grfico de Mtricas por Modelo

# %%
# Create summary visualizations to compare models
# Convert columns to numerical for visualization
numeric_cols = ["RMSE", "MAE", "MAPE(%)", "R", "NSE", "Pearson_Corr"]
df_plot = df_comparative.copy()

for col in numeric_cols:
    if col in df_plot.columns:
        df_plot[col] = pd.to_numeric(df_plot[col], errors='coerce')

# Create a composite model for grouping and average across horizons
df_plot['Modelo_Compuesto'] = df_plot['Modelo'] + '_' + df_plot['Rama']
# Handle "Meta-Modelo" and "Ensemble_Average" specifically
df_plot.loc[df_plot['Modelo'] == 'Meta-Modelo', 'Modelo_Compuesto'] = 'Meta-Modelo'
df_plot.loc[df_plot['Modelo'] == 'Ensemble_Average', 'Modelo_Compuesto'] = 'Ensemble_Average'


# Calculate averages per model (across horizons and dates)
df_avg = df_plot.groupby('Modelo_Compuesto')[numeric_cols].mean(numeric_only=True).reset_index()

# Drop entries with NaN in Modelo_Compuesto if generated
df_avg = df_avg.dropna(subset=['Modelo_Compuesto'])

# Order the models for consistent plotting
model_order = [f"{m}_{b}" for m in METHODS for b in BRANCHES] + ["Ensemble_Average", "Meta-Modelo"]
# Ensure Modelo_Compuesto is string type for plotting
df_avg['Modelo_Compuesto'] = df_avg['Modelo_Compuesto'].astype(str)
df_avg['Modelo_Compuesto'] = pd.Categorical(df_avg['Modelo_Compuesto'], categories=model_order, ordered=True)
df_avg = df_avg.sort_values('Modelo_Compuesto').reset_index(drop=True)


# 1. Bar chart for RMSE and MAE
plt.figure(figsize=(16, 9)) 
width = 0.35
x = np.arange(len(df_avg))

# Convert `Modelo_Compuesto` back to list of strings for plotting x-labels
plt.bar(x - width/2, df_avg['RMSE'], width, label='RMSE', color='skyblue')
plt.bar(x + width/2, df_avg['MAE'], width, label='MAE', color='salmon')

plt.xlabel('Modelo')
plt.ylabel('Valor')
plt.title('RMSE y MAE por Modelo (Promedio sobre Horizontes)')
plt.xticks(x, df_avg['Modelo_Compuesto'].tolist(), rotation=45, ha='right') # Pass as list of strings
plt.legend()
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 2. Bar chart for MAPE
plt.figure(figsize=(16, 7)) 
plt.bar(df_avg['Modelo_Compuesto'].tolist(), df_avg['MAPE(%)'], color='lightgreen') # Pass as list of strings
plt.xlabel('Modelo')
plt.ylabel('MAPE (%)')
plt.title('Error Porcentual Absoluto Medio (MAPE) por Modelo (Promedio sobre Horizontes)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# 3. Bar chart for R and NSE
plt.figure(figsize=(16, 8)) 
width = 0.35
x = np.arange(len(df_avg))

plt.bar(x - width/2, df_avg['R'], width, label='R', color='lightblue')
plt.bar(x + width/2, df_avg['NSE'], width, label='NSE', color='lightgreen')

plt.xlabel('Modelo')
plt.ylabel('Valor')
plt.title('Coeficientes R y NSE por Modelo (Promedio sobre Horizontes)')
plt.xticks(x, df_avg['Modelo_Compuesto'].tolist(), rotation=45, ha='right') # Pass as list of strings
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.axhline(y=1, color='green', linestyle='--', alpha=0.5) 
plt.axhline(y=0, color='red', linestyle='--', alpha=0.5)   
plt.show()

# %% [markdown]
# ## 10) Guardar modelos entrenados

# %%
n_cells = ds_full.latitude.size * ds_full.longitude.size
INPUT_WINDOW = 60 # Asegurar que INPUT_WINDOW est disponible

for name in histories.keys():
    model_path = MODELS_OUTPUT / f"{name}_final.pt"
    branch = name.split("_")[-1]
    
    # Calculate current_n_feats dynamically, same as in the training section
    if branch == "low":
        current_n_feats = n_cells + 2 
    else:
        current_n_feats = n_cells 

    # Re-instantiate the model with the correct parameters to load the state_dict
    if branch == "low":
        model = GRUEncoderDecoder(current_n_feats, 128, OUTPUT_HORIZON, n_cells, dropout=0.0)
    else:
        model = TransformerModel(current_n_feats, OUTPUT_HORIZON, n_cells, dropout=0.0) # Ensure dropout is passed
    
    # Load the model and save it to the new location
    try:
        load_path = MODELS_OUTPUT / f"best_base_{name}.pt"
        model.load_state_dict(torch.load(load_path, weights_only=False)) 
        torch.save(model.state_dict(), model_path)
        logger.info(f"Base model saved: {model_path} with n_feats={current_n_feats}")
    except Exception as e:
        logger.error(f"Error saving model {name}: {e}")

# Only save meta-model if it was trainable
if meta_model_trainable:
    logger.info(f"Final meta-model saved to {MODELS_OUTPUT / 'meta_model_final.pt'}")
else:
    logger.warning("Meta-model was not trainable. Skipping saving meta-model.")
```
